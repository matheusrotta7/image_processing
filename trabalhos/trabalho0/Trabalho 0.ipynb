```python
from scipy import misc;
from scipy import ndimage;
import numpy as np;
import matplotlib.pyplot as plt;
import imageio;
```

```python
#open image file and store it in a numpy array
img = imageio.imread('images/city.png');
```

```python
#evaluate img
print(img);
print("\n**Image shape: " + str(img.shape));
plt.imshow(img, cmap='gray')
```

<h3>Transformao de Intensidade</h3>

```python
#(i) Obter o negativo da imagem
negative_img = 255 - img;
print(negative_img);
plt.imshow(negative_img, cmap='gray');
```

```python
#(2) Converter o intervalo de intensidades para [100, 200]
# img = np.array(img)

new_img = np.ones((512, 512))
print(new_img)
print(new_img.shape)
print(new_img[0][0])
for i in range(0,512):
    for j in range (0,512):
        new_img[i][j] = (img[i][j] * 100 + 25500)//255;
print(new_img)
print(new_img.shape)
plt.imshow(new_img, cmap='gray')
```
